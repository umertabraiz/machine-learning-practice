```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.read_csv('/content/drive/MyDrive/Salary_dataset.csv')

plt.xlabel("YearsExperience", color = "red")
plt.ylabel("Salary", color = "red")
plt.scatter(df.YearsExperience, df.Salary)
```

```python
lr = linear_model.LinearRegression()
lr.fit(df[["YearsExperience"]], df.Salary)

lr.predict([[10.5]]) # 124072.8083418

# How to check the prediction with the formula
lr.score(df[["YearsExperience"]], df.Salary) # 0.9569566641435086
lr.coef_ # 9449.96232146
lr.intercept_ # 24848.203966523208


predicted = 9449.96232146*10.5 + 24848.203966523208 # y =wx + b where w is coef & x is value of x & b is intercept
print(predicted)
```

```python
plt.xlabel("YearsExperience", color = "red")
plt.ylabel("Salary", color = "red")
plt.scatter(df.YearsExperience, df.Salary)
plt.plot(df.YearsExperience, lr.predict(df[["YearsExperience"]]), color = 'red')
```
